```python
import fuzzywuzzy.fuzz as fuzz

def fuzzy_similarity(entity1, entity2):
    
    similarity1 = fuzz.token_sort_ratio(entity1['attribute1'], entity2['attribute1'])
    similarity2 = fuzz.partial_ratio(entity1['attribute2'], entity2['attribute2'])
    similarity3 = fuzz.token_set_ratio(entity1['attribute3'], entity2['attribute3'])
    
    overall_similarity = (similarity1 * weight1 + similarity2 * weight2 + similarity3 * weight3) / (weight1 + weight2 + weight3)
    
    return overall_similarity
```

```python
lf=lfa1

import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

def find_duplicate_vendors(vendors_df, threshold):
    duplicates = []
    lf = vendors_df.copy()
    lf['NAME1'] = lf['NAME1'].astype(str)
    vendor_names = lf['NAME1'].unique().tolist()
    columns = ['Vendor 1', 'Vendor 2', 'Score']
    df_duplicates = pd.DataFrame(columns=columns)

    for i, name in enumerate(vendor_names):
        # Compare the current name with the remaining names
        matches = process.extract(name, vendor_names[i+1:], scorer=fuzz.ratio)

        # Check if any match exceeds the threshold
        for match, score in matches:
            if score >= threshold:
                duplicates.append((name, match))
                df_duplicates = df_duplicates.append({'Vendor 1': name, 'Vendor 2': match, 'Score': score}, ignore_index=True)

    return duplicates, df_duplicates

# Example usage
threshold = 95
duplicate_vendors, df_duplicate_vendors = find_duplicate_vendors(lf, threshold)

# Print the duplicate vendor pairs
print("Duplicate Vendor Accounts:")
for vendor1, vendor2 in duplicate_vendors:
    print(vendor1, "--", vendor2)

# Print the dataframe with duplicate vendor pairs
print("Duplicate Vendor Accounts DataFrame:")
print(df_duplicate_vendors)
```
